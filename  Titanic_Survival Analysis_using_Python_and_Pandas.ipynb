Loading the Titanic dataset EDA using Pandas

Code Explanation:  
I used Pandas to load the Titanic dataset directly from GitHub. The `read_csv` function reads the CSV file and stores it in a DataFrame named `df`, which works like a table with rows and columns. Displaying `df.head()` helped me confirm that the data loaded correctly and showed the column names and first few passenger records.

```python
import pandas as pd

# Load the Titanic dataset from the URL
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Check that it loaded
print(df.head())
```

Question 2

Performing EDA on the Titanic Dataset


Code Explanation:

After loading the data, I explored its basic structure. The `shape` showed that the dataset contains 891 passengers and 12 features. Using `info()` revealed that some columns like `Age` and `Cabin` have missing values, and it also showed the data type for each column.  

The `describe()` function provided a statistical summary of the numerical columns, such as the average age, fare, and passenger ID range. From `isnull().sum()`, I observed that the `Cabin` column has the most missing values, followed by `Age` and `Embarked`. This step is important because missing data can affect analysis later.

```python

# Calculate the overall survival rate
print("\nOverall survival rate:")
print(df['Survived'].mean())

# Compare survival rate by gender
print("\nSurvival rate by gender:")
print(df.groupby('Sex')['Survived'].mean())

# Compare survival rate by passenger class
print("\nSurvival rate by class:")
print(df.groupby('Pclass')['Survived'].mean())
```

Explanation:
I analyzed the `Survived` column to understand patterns in survival. The overall survival rate was about 38%, meaning roughly 4 out of 10 passengers survived.  

When grouped by gender, females had a survival rate of about 74%, while males had around 19%. This suggests gender played a major role in survival.  

Grouping by passenger class showed that passengers in 1st class had the highest survival rate, followed by 2nd class, and then 3rd class. This indicates that socioeconomic status also influenced the chances of survival.
